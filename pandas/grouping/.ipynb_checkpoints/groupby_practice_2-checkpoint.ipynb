```python
import pandas as pd
```

```python
dataset_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(dataset_url, sep='|')
users.head()
```

```python
occupation_grp = users.groupby('occupation')
occupation_grp['age'].mean()
```

```python
# from collections import Counter
# def cal_male_count(series):
#     c = 0
#     for i in series:
#         if i == "M":
#             c += 1
#     return c

# def cal_female_count(series):
#     c = 0
#     for i in series:
#         if i == "F":
#             c += 1
#     return c

def cal_gender_count(series, gender):
    c = 0
    for i in series:
        if i == gender:
            c += 1
    return c
    
gender_df = occupation_grp.agg(Male = (('gender', lambda x: cal_gender_count(x, 'M'))),
                   Female = (('gender', lambda x: cal_gender_count(x, "F"))),
                   Total = (('gender', 'count'))
                  )
gender_df['Male Ratio'] = gender_df['Male'] / gender_df['Total']
gender_df.sort_values(by='Male Ratio', ascending=False) 
```

```python
# gender_df['Female Ratio'] = gender_df["Female"] / gender_df['Total']
# gender_df


# Group by occupation and gender, then count the occurrences
occupation_gender_counts = users.groupby(['occupation', 'gender']).size().reset_index()
gender_df = occupation_gender_counts.pivot(index='occupation', columns=['gender'], values=0)
# occupation_gender_counts
gender_df.fillna(0, inplace=True)
gender_df['Total'] = gender_df['F'] + gender_df["M"]
gender_df['Male Ratio'] = gender_df['M'] / gender_df['Total']
gender_df['Female Ratio'] = gender_df['F'] / gender_df['Total']
gender_df
```
