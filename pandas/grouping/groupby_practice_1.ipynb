```python
import pandas as pd
```

```python
dataset_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
drinks = pd.read_csv(dataset_url)
drinks.head()
```

```python
continent_grps = drinks.groupby('continent')
continent_grps['beer_servings'].mean().sort_values(ascending=False).head(1)
```

```python
continent_grps['wine_servings'].agg(['mean', 'median', 'count', 'std'])
continent_grps['wine_servings'].describe()
```
