```python
import pandas as pd
```

```python
dataset_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv"
```

```python
apple = pd.read_csv(dataset_url)
# apple.head(30)
```

```python
apple.info()
```

```python
apple.Date = pd.to_datetime(apple.Date)
apple.Date.dtype
```

```python
apple.set_index('Date', inplace=True)
```

```python
apple.index.nunique() == apple.shape[0]
```

```python
apple = apple.sort_index(ascending=True)
apple
```
