```python
import pandas as pd
import numpy as np
```

```python
dataset_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
orders = pd.read_csv(dataset_url, sep='\t')
```

```python
orders.head()
```

```python
orders.loc[0:5] # end index is implicit in loc
```

```python
orders.iloc[0:5] # end index is explicit in iloc
```

```python
orders.loc[0:5, 'quantity'] # returns a series if a single columns is accessed
```

```python
orders.loc[0:5:2, ['quantity']] #returns a df if a list of columns are accessed
```

```python
orders.loc[:10, ['quantity', 'item_name']]
```
