```python
import pandas as pd
import numpy as np
```

```python
car1_dataset_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv"
car2_dataset_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv"
car1 = pd.read_csv(car1_dataset_url)
car2 = pd.read_csv(car2_dataset_url)
```

```python
car1.head()
```

```python
car2.head()
```

```python
car1.drop(['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13' ], axis=1, inplace=True)
car1.head()
```

```python
car1.shape
```

```python
car2.shape
```

```python
cars = pd.concat([car1, car2], axis=0)
# cars = car1.append(car2)
cars.head()
```

```python
owners = np.random.randint(15000, 73000, size=(cars.shape[0]))
owner_series = pd.Series(owners, name="Owners")
owner_series
```

```python
cars['Owners'] = owner_series
```
