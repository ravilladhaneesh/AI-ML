```python
import pandas as pd
import numpy as np
```

```python
dataset_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wines = pd.read_csv(dataset_url)
wines
```

```python
# first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns
wines = wines.iloc[:, [2,3,5,6,8,10,12]]
wines
```

```python
wines.columns = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
wines.head()
```

```python
wines.iloc[:3, 0] = np.nan
wines
```

```python
wines.loc[2:3, "magnesium"] = np.nan
wines
```

```python
# wines['alcohol'] = wines['alcohol'].fillna(10)
# wines['magnesium'] = wines['magnesium'].fillna(100)

# wines.alcohol = wines.alcohol.replace({np.nan: 10})

filt = wines.alcohol.isna()
wines.loc[filt, 'alcohol'] = 10

filt = wines.magnesium.isna()
wines.loc[filt, 'magnesium'] = 100

wines.head()
```
