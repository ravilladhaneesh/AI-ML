```python
data = pd.read_csv("files/custom_data.csv")
data
```

```python
from collections import Counter
```

```python
language_counter = Counter()

for languages in data['LanguageHaveWorkedWith']:
    language_counter.update(languages.split(';'))
sorted_language_counter = dict(sorted(language_counter.items(), key=lambda x: x[1]))

language = [lang for lang in sorted_language_counter]
lang_value = [sorted_language_counter[lang] for lang in sorted_language_counter]
```

```python
plt.subplots(figsize=(10, 15)) # width = 10 inches and tall = 15 inches
plt.barh(language, lang_value)
plt.xlabel("people count")
plt.title("People count knowing the langauage")
plt.tight_layout()
plt.show()
```
