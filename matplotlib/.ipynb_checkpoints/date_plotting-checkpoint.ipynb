```python
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_date
from datetime import datetime, timedelta
import numpy as np
```

```python
def generate_dates(total_days):
    dates = [datetime.now().date()]
    for i in range(1, total_days):
        next_day = timedelta(1) + dates[-1] 
        dates.append(next_day)
    return dates

dates = generate_dates(10)
y = np.random.randint(30, size=(10,))
```

```python
print(y)
```

```python
date_format = mpl_date.DateFormatter('%dth, %b-%y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.plot_date(dates, y, linestyle='-', label='Count on dates')
plt.gcf().autofmt_xdate()
# plt.xticks(rotation=45)
plt.xlabel('Dates')
plt.ylabel('Count')
plt.legend()
plt.show()
```
