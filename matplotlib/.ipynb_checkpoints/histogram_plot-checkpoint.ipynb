```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('ggplot')
```

```python
df = pd.read_csv('../files/data.csv')
```

```python
ages = [18, 20, 21, 22, 25, 27, 29, 31, 34, 34, 39, 41, 44, 47, 51, 55]
bins = [10, 20, 30, 40, 50, 60, 70]
```

```python
plt.hist(ages, bins=bins, edgecolor='black', color='red', alpha=0.5, label='Age count')
plt.legend()
plt.xlabel('Ages of Respondent')
plt.ylabel('Count')
plt.show()
```

```python
experience = df['WorkExp'].dropna()
bins = [20, 30, 40, 50, 60, 70, 80]

mean_experience = experience.mean()
```

```python
plt.hist(experience, bins=bins, edgecolor='black', color='blue', alpha=0.25, label='Experience count') #, log=True

# 20 for starting bin(assume working after age=20)
plt.axvline(mean_experience + 20, color='black', linewidth=2, label='mean experience') 
plt.xlabel('Experience')
plt.ylabel('count')
plt.legend()
plt.show()
```
