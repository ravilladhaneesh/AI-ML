```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.random.randint(20, size=(20,))
y = np.random.randint(20, size=(20,))
c = np.random.randint(10, size=(20,))

print(x)
print(y)
print(c)
```

```python
plt.scatter(x, y, c=c, cmap='Oranges')# , marker='x'
cbar = plt.colorbar()
cbar.set_label('Range')
# plt.xscale('log')
plt.tight_layout()
plt.show()
```
